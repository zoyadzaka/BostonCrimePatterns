## Atribut pada Dataset :

1. COMPNOS - Nomor unik untuk setiap laporan kejadian kriminal.
2. NatureCode - Kode kategori kejadian kriminal.
3. INCIDENT_TYPE_DESCRIPTION - Deskripsi tipe insiden kriminal
4. MAIN_CRIMECODE - Kode utama kejahatan, mengelompokkan insiden berdasarkan kategori tertentu
5. REPORTDISTRICT - Distrik tempat kejadian dilaporkan.
6. REPORTINGAREA - Area pelaporan kejadian dalam distrik tertentu.
7. FROMDATE - Tanggal dan waktu saat insiden terjadi.
8. WEAPONTYPE - Jenis senjata yang digunakan dalam kejadian
9. Shooting - Indikator apakah kejadian melibatkan penembakan ("Yes" atau "No").
10. DOMESTIC - Indikator apakah kejahatan termasuk dalam kategori kekerasan domestik.
11. SHIFT - Waktu kejadian berdasarkan shift kerja polisi (seperti "Day", "Last", atau "First").
12. Year - Tahun terjadinya insiden.
13. Month - Bulan kejadian insiden (1-12).
14. DAY_WEEK - Hari kejadian dalam seminggu (misalnya "Sunday", "Monday").
15. UCRPART - Kategori kejahatan berdasarkan sistem pelaporan kejahatan (seperti "Part One", "Part Two", "Part Three").
16. X - Koordinat X lokasi kejadian.
17. Y - Koordinat Y lokasi kejadian.
18. STREETNAME - Nama jalan tempat insiden terjadi.
19. XSTREETNAME - Nama persimpangan jalan jika tersedia.
20. Location - Titik koordinat geografis (latitude, longitude) kejadian.

##Step 2 :  Data Cleaning

### Handling Duplicated Data


```python
#Mengecek total  data yang duplikat
df.duplicated().sum() #-> Terdapat 1 duplikat data
```

```python
#Melihat data yang mana saja yang duplikat
df[df.duplicated(keep = False)]
```

```python
#Menghapus kolom duplikat
df.drop_duplicates(inplace=True)
```

```python
#Memastikan kembali apakah masih ada data duplikat
df.duplicated().sum() # -> Sudah tidak ada data duplikat karena total data duplikat sudah 0
```

###Handling Missing Value

```python
#Cek apakah ada data yang kosong
df.isna().sum()
```

#####Drop Attribute, karena XSTREETNAME yaitu atribut untuk menyimpan data nama persimpangan jalan memiliki missing values yang sangat besar hingga 2.715 baris yang tidak memiliki isi dari total 3000 baris, kita drop saja karena nama jalan masih dapat menggunakan atribut 'STREETNAME'

```python
#Drop Attribute
df.drop(columns=['XSTREETNAME'], inplace=True)
```

####Untuk X dan Y karena berkaitan dengan koordinat lokasi, kami tidak akan handling menggunakan modus karena bisa membuat data menjadi missleading dengan informasi koordinat yang salah

####Dengan demikian kami akan menghapus baris yang koordinat X dan Y nya kosong

```python
#Drop Missing Values di setiap row
df.dropna(subset=['X', 'Y'], inplace=True)

# Cek apakah masih ada missing values
print(df.isna().sum())
```

####Untuk Weapon Type dan Shift karena tipe datanya adalah kategorikal, kita dapat mengisinya dengan nilai modus dari tiap atribut, sedangkan untuk COMPNOS akan kita isi missing valuenya dengan median karena berupa numerik

```python
#Mengisi missing value pada Weapon Type
df['WEAPONTYPE'].fillna(df['WEAPONTYPE'].mode()[0], inplace =True)

#Mengisi missing value pada Shift
df['SHIFT'].fillna(df['SHIFT'].mode()[0], inplace =True)

#Mengisi missing value pada COMPNOS
df['COMPNOS'].fillna(df['COMPNOS'].median(), inplace=True)

#Cek apakah masih ada missing values
print(df.isna().sum())
```

#### Data sudah bersih dari missing values dan duplikat!

## Step 3 : Explore Data

#### 1. Tren kejahatan dari waktu ke waktu

```python
# Menghitung jumlah kejahatan per tahun
crime_per_year = df['Year'].value_counts().sort_index()

#Membuat plot
plt.figure(
    figsize=(10,6)
)
sns.lineplot(x=crime_per_year.index,
             y= crime_per_year.values,
             marker='o',
             linestyle='-',
             linewidth=2.5,
             markersize=8,
             color='#FF6F61'
)

plt.fill_between(crime_per_year.index, crime_per_year.values, alpha=0.2, color="#FF6F61") #shadow dibawah garis

# Judul
plt.title("Tren Kejahatan per Tahun", fontsize=16, fontweight='bold', color="#333333")

# Label sumbu X dan Y
plt.xlabel("Tahun", fontsize=14, fontweight='bold', color="#444444")
plt.ylabel("Jumlah Kejahatan", fontsize=14, fontweight='bold', color="#444444")

# Menampilkan plot
plt.show()
```

####2. Rata-rata Kejahatan Berdasarkan Distrik


```python
# Hitung jumlah kejahatan per distrik
crime_per_district = df['REPTDISTRICT'].value_counts()

# Buat plot
plt.figure(figsize=(12, 7))
ax = sns.barplot(
    y=crime_per_district.index,
    x=crime_per_district.values,
    palette="Reds_r",
    edgecolor="black",
    linewidth=1
)

# Tambahkan label angka di setiap bar
for container in ax.containers:
    ax.bar_label(container, fmt='%d', padding=3, fontsize=12, color='black', weight='bold')

#jJudul
plt.title("Jumlah Kejahatan Berdasarkan Distrik", fontsize=16, fontweight='bold')
plt.xlabel("Jumlah Kejahatan", fontsize=14)
plt.ylabel("Distrik", fontsize=14)


# Tampilkan plot
plt.show()
```

####3. Jenis kejahatan yang paling umum

```python
# Mengambil 10 jenis kejahatan terbanyak
crime_counts = df['INCIDENT_TYPE_DESCRIPTION'].value_counts().head(10)

# Membuat figure
plt.figure(figsize=(12, 6))
ax = sns.barplot(x=crime_counts.values, y=crime_counts.index, palette="Reds_r", edgecolor='black')

# Menambahkan label angka pada tiap bar
for container in ax.containers:
    ax.bar_label(container, fmt='%d', padding=3, fontsize=12, color='black', weight='bold')

# Menambahkan judul dan label
plt.title("Top 10 Jenis Kejahatan di Boston", fontsize=14, fontweight='bold')
plt.xlabel("Jumlah Kejadian", fontsize=12)
plt.ylabel("Jenis Kejahatan", fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```

####4. Hubungan antara Shift Kerja Polisi dengan Jumlah Kejahatan

```python
# Preprocessing
df['FROMDATE'] = pd.to_datetime(df['FROMDATE'], errors='coerce')
df['Month'] = df['FROMDATE'].dt.month
df['Year'] = df['FROMDATE'].dt.year
df['DAY_WEEK'] = df['FROMDATE'].dt.day_name()  # Menambahkan kolom hari

# Buat pivot table jumlah kejadian berdasarkan hari dan bulan
pivot_table = pd.pivot_table(df, index='DAY_WEEK', columns='Month',
                              values='INCIDENT_TYPE_DESCRIPTION', aggfunc='count')

# Urutan hari
order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Plot
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table.loc[order], cmap='Reds_r', annot=True, fmt='.0f')
plt.title('Heatmap Jumlah Kejahatan per Hari dan Bulan')
plt.xlabel('Bulan')
plt.ylabel('Hari')
plt.tight_layout()
plt.show()
```

###Heatmap menunjukkan distribusi kejahatan berdasarkan shift kerja polisi dan hari dalam seminggu. Terlihat bahwa kejahatan lebih sering terjadi pada shift malam, terutama di akhir pekan. Hal ini mengindikasikan peningkatan aktivitas kriminal saat pengawasan mungkin lebih terbatas. Polisi dapat menggunakan informasi ini untuk mengoptimalkan penempatan personel pada waktu-waktu rawan

```python
shift_counts = df['SHIFT'].dropna().value_counts()
pastel_colors = ['#FFB3BA', '#D7BDE2', '#ADD8E6']
plt.figure(figsize=(6, 6))
plt.pie(shift_counts, labels=shift_counts.index, autopct='%1.1f%%', startangle=90,colors=pastel_colors)
plt.title('Distribusi Kejahatan Berdasarkan Waktu (Shift)')
plt.tight_layout()
plt.show()
```

```python
weapon_counts = df['WEAPONTYPE'].dropna().value_counts()
plt.figure(figsize=(10, 8))

# Hitung persentase
percentages = weapon_counts / weapon_counts.sum() * 100
labels_with_percent = [f"{label} ({p:.2f}%)" for label, p in zip(weapon_counts.index, percentages)]

# Buat pie chart TANPA autopct (persentase dihilangkan dari chart)
wedges = plt.pie(
    weapon_counts,
    startangle=140,
    radius=1.2,
    wedgeprops={'width': 0.6}  # Untuk donut chart
)[0]

# Buat donut (lingkaran putih di tengah)
plt.gca().add_artist(plt.Circle((0, 0), 0.5, color='white'))

# Buat legend dengan label yang sudah include persentase
plt.legend(
    wedges,
    labels_with_percent,  # Gunakan label yang sudah ada persentasenya
    title="Tipe Senjata (Persentase)",
    loc="center left",
    bbox_to_anchor=(1, 0.5),
    fontsize=9
)

plt.title('Distribusi Tipe Senjata', pad=20)
plt.tight_layout()
plt.show()
```
